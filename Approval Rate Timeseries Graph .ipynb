```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
hmda = pd.read_csv('cleaned.csv')
```

```python
hmda
```

```python
race_mapping = {5: 'White', 3: 'Black or African American', 2: 'Asian', 1: 'American Indian or Alaska Native', 4: 'Native Hawaiian or Other Pacific Islander'}
hmda['action_taken'] = hmda['action_taken'].apply(lambda x: 1 if x in [1, 2] else 0)
hmda['applicant_race_1'] = hmda['applicant_race_1'].map(race_mapping)
hmda.describe()
approval_rates = hmda.groupby(['as_of_year', 'applicant_race_1'])
```

```python
# approval_rates_by_race = hmda.groupby(['as_of_year', 'applicant_race_1'])['action_taken'].mean() * 100
approval_rates_by_race = hmda.groupby(['applicant_race_1'])['action_taken'].mean() * 100
```

```python
approval_rates_by_race_df = approval_rates_by_race.reset_index() 
approval_rates_by_race_df.columns = ['Race', 'Approval Rate (%)']  
approval_rates_by_race_df
```

```python
approval_rates_by_race_year = hmda.groupby(['as_of_year', 'applicant_race_1'])['action_taken'].mean() * 100
```

```python
approval_rates_by_race_year_df = approval_rates_by_race_year.reset_index() 
approval_rates_by_race_year_df.columns = ['Year', 'Race', 'Approval Rate (%)']
approval_rates_by_race_year_df
```

```python
time_series_df = approval_rates_by_race_year_df.pivot(index='Year', columns='Race', values='Approval Rate (%)')
```

```python
plt.figure(figsize=(20, 20))
time_series_df.plot(marker='o')

plt.title('Loan Approval Rates by Race Over Time', fontsize=12)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Approval Rate (%)', fontsize=12)
plt.legend(title="Race", bbox_to_anchor=(1.05, 1.0), loc='upper left')
plt.grid(True)
plt.show()
```
